```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math 
import os
```

```python
spec_pion_exp=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\spec_pi_exp.dat", sep="\t", header=None, names=["pt", "pt_max", "pt_min","spec","chyba","chybaminus","sys","sysminus"])
spec_proton_exp=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\spec_p_exp.dat", sep="\t", header=None, names=["pt", "pt_max", "pt_min","spec","chyba","chybaminus","sys","sysminus"])
spec_kaon_exp=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\spec_k_exp.dat", sep="\t", header=None, names=["pt", "pt_max", "pt_min","spec","chyba","chybaminus","sys","sysminus"])

spec_hadrons=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\spectrum.dat", sep="\t", header=None, names=["pt", "pion","kaon","proton"])
antispec_hadrons=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\antispectrum.dat", sep="\t", header=None, names=["pt", "pion","kaon","proton"])

antispec_pion_exp=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\spec_barpi_exp.dat", sep="\t", header=None, names=["pt", "pt_max", "pt_min","spec","chyba","chybaminus","sys","sysminus"])
antispec_proton_exp=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\spec_barp_exp.dat", sep="\t", header=None, names=["pt", "pt_max", "pt_min","spec","chyba","chybaminus","sys","sysminus"])
antispec_kaon_exp=pd.read_csv("C:\\Users\\LPO\\Desktop\\vyzkumak\\spec_bark_exp.dat", sep="\t", header=None, names=["pt", "pt_max", "pt_min","spec","chyba","chybaminus","sys","sysminus"])

```

```python
fig, axes = plt.subplots(sharex=True, figsize=(13,10))

axes.errorbar(spec_pion_exp["pt"],spec_pion_exp["spec"], yerr =(spec_pion_exp["chyba"]), fmt='o',ecolor='r',color='black',elinewidth=2,capsize=4, label="CERN Alice data")
axes.errorbar(spec_proton_exp["pt"],spec_proton_exp["spec"]/5, yerr =(spec_proton_exp["chyba"]), fmt='o',ecolor='r',color='black',elinewidth=2,capsize=4)
axes.errorbar(spec_kaon_exp["pt"],spec_kaon_exp["spec"]/2, yerr =(spec_kaon_exp["chyba"]), fmt='o',ecolor='r',color='black',elinewidth=2,capsize=4)

axes.plot(spec_hadrons["pt"],spec_hadrons["pion"], color='blue',linewidth=2, label="Trento IS Pb+Pb 2.76TeV 20-30%")
axes.plot(spec_hadrons["pt"],spec_hadrons["kaon"]/2, color='blue',linewidth=2)
axes.plot(spec_hadrons["pt"],spec_hadrons["proton"]/5,color='blue',linewidth=2)


axes.set_yscale('log')
axes.grid(True)
#axes.set_xlim([0.2, 2.1])
#axes.set_ylim([0.01, 120])
axes.annotate('Pb+Pb 2.76TeV 20-30%', (1.5, 10),fontsize=25)
axes.annotate('$\pi^+$', (0.4, 200),fontsize=20)
axes.annotate('$K^+/2$', (0.25, 30),fontsize=20)
axes.annotate('$p/5$', (0.3, 1.3),fontsize=20)
axes.set_xlabel('$p_T$[GeV]',fontsize=20)
axes.set_ylabel('$d^2N/(2πp_Tdp_Tdy)$[$GeV^{-2}$]',fontsize=20)
axes.tick_params(axis='x', labelsize=20)
axes.tick_params(axis='y', labelsize=20)
axes.legend(loc=0,fontsize=15)
fig.savefig("spectrum_276_PbPb.pdf")
```
